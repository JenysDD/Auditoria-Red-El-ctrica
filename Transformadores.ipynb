```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

try:
  print("---INICIANDO PROYECTO DE INTEGRACION----")

  # 1. Base de datos de la red (Pandas)
  db = {
      "transfo" : ["T1", "T2", "T3"],
      "nom" : [100, 250, 400],
      "real": [85, 270, 340]
  }
  df = pd.DataFrame(db)

  # 2. Calculo masivo en paralelo (NumPy)
  # Covertimos a arrays y calculamos el % de carga real
  nom_arr = np.array(df["nom"])
  real_arr = np.array(df["real"])
  carga_porc = (real_arr/nom_arr) * 100

  # Acoplamos el calculo a la tabla de Pandas
  df["carga_%"] = carga_porc
  print("\n 〽 TABLA DE CONTROL DE SUBESTACIONES:")
  print(df)
  print("-" * 50)

  # 3. Ploteo de gala ( Seaborn + matplotib)
  sns.set_theme(style="whitegrid")

  # Creamos el Dashboard de barras con la paleta tecnologica ajustada
  sns.barplot(x="transfo", y="carga_%", data=df, palette="Blues_d", hue="transfo", legend=False)

  # Añadimos la linea critica del 100% de carga segun ingenieria

  plt.axhline(100, color="red", linestyle="dashed", label="Limite critico(100%)")

  # Etiquetas del plano visual
  plt.title("Auditoria de carga en transformadores")
  plt.xlabel("Equipos en Red")
  plt.ylabel("% Carga Real")

  print("💫 Desplegando interfaz grafica en la pc")
  plt.show()


except Exception as e:
  print(f"❌ Cortocircuito en el software: {e}")


```
